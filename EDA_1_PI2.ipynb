Para saber el porcentaje de datos nulos con exactitud, se crea la siguiente función que explica mejor las métricas en datos nulos.

```python
def DataType(df):
    dfDict = {
        "name": [], "data_type": [], "not_null_%": [], "null_%": [], "null": []
    }
    for column in df.columns:
        notNull = (df[column].count() / len(df)) * 100
        dfDict["name"].append(column)
        dfDict["data_type"].append(df[column].apply(type).unique())
        dfDict["not_null_%"].append(round(notNull, 2))
        dfDict["null_%"].append(round(100-notNull, 2))
        dfDict["null"].append(df[column].isnull().sum())
    dfDataType=pd.DataFrame(dfDict)
    return dfDataType
```

Se puede ver que cuando no hay datos tanto en 'altura' como en 'cruce', el tipo de calle donde ocurrió el hecho es principalmente 'AUTOPISTA'. Para corroborar esto, se crea una función que nos arroje el porcentaje y la cantidad de los datos de la columna 'tipo de calle'.

```python
def DataPorcentaje(df, column):
    counts = df[column].value_counts()
    percentages = round(100 * counts / len(df), 2)
    dfResults = pd.DataFrame({
        "Cantidad":counts,
        "Porcentaje":percentages
    })
    return dfResults
```

### Variables Geográficas
Verificamos los datos geográficos ya que son parte importante para un análisis de accidentes de tránsito; así que traemos a observación los datos de las columnas 'xy_(caba)', 'pos_x', y 'pos_y', que contienen respectivamente datos de: geocodificación plana, longitud con separador punto y latitud con separador punto. 

Vemos que podemos extraer coordenadas de la columna 'xy_(caba)', así que se utiliza una función para dicho fin:

```python
# Esta función utiliza una expresión regular (re.findall) para buscar todas las coincidencias de números decimales (que tienen un punto como separador decimal) en el texto ingresado. 
# Estas coincidencias se almacenan en la lista coordenadas.
# return float(coordenadas[0]), float(coordenadas[1]): Si se encontraron dos coordenadas, se convierten en números de punto flotante (float) y se devuelven como una tupla de dos valores.
# Luego, se aplica esta función a la columna 'XY (CABA)' de df_siniestros utilizando el método .map(). 
# Esto se hace para cada fila de la columna, y los resultados se almacenan en dos nuevas columnas, 'coord_x' y 'coord_y', utilizando zip(*...).

def extraer_coordenadas(texto):
    # Utilizamos una expresión regular para extraer las coordenadas
    coordenadas = re.findall(r'\d+\.\d+', texto)
    if len(coordenadas) == 2:
        return float(coordenadas[0]), float(coordenadas[1])
    else:
        return None, None

# Aplicamos la función a la columna 'xt (caba)'
dfHechos['coord_x'], dfHechos['coord_y'] = zip(*dfHechos['xy (caba)'].map(extraer_coordenadas))

# Filtrar los registros que no tienen coordenadas válidas
dfHechos = dfHechos[(dfHechos['coord_x'].notnull()) & (dfHechos['coord_y'].notnull())]
```
